```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.gridspec as gridspec

%matplotlib notebook

fig = plt.figure()
gspec = gridspec.GridSpec(2, 2)

# generate 4 random variables from the random, gamma, exponential, and uniform distributions

top_right = plt.subplot(gspec[0, 1:])
top_left = plt.subplot(gspec[0, 0])
lower_left = plt.subplot(gspec[1:, 0])
lower_right = plt.subplot(gspec[1:, 1:])

x1 = np.random.normal(-2.5, 1, 10000)
x2 = np.random.gamma(2, 1.5, 10000)
x3 = np.random.exponential(2, 10000)
x4 = np.random.uniform(14,20, 10000)


top_right.hist(x2, normed=True, bins=100, alpha=0.5, color='green')
top_left.hist(x1, normed=True, bins=100, alpha=0.5, color='red')
lower_left.hist(x3, normed=True, bins=100, alpha=0.5, color='brown')
lower_right.hist(x4, normed=True, bins=100, alpha=0.5, color='blue')

top_right.set_title('Gamma')
top_left.set_title('Random')
lower_left.set_title('Exponential')
lower_right.set_title('Uniform')

fig.subplots_adjust(hspace=0.5)
```
